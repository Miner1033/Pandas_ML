```python
df.isnull()#`df.isnull()` returns a DataFrame of the same shape as `df`, with `True` at each position where there is an `NaN` (missing) value and `False` where there is not. This is useful for identifying missing values in the DataFrame.
```

```python
df.isnull().sum()#`df.isnull().sum()` returns the count of `NaN` (missing) values in each column of the DataFrame `df`, giving a quick summary of missing data per column.
```

```python
df.isnull().sum().sum()#`df.isnull().sum().sum()` returns the total count of `NaN` (missing) values across the entire DataFrame `df`, providing a single number that represents all missing entries.
```

```python
df.notnull()#`df.notnull()` returns a DataFrame with `True` for non-missing values and `False` for `NaN` (missing) values in `df`.
```

```python
df.notnull().sum()#`df.notnull().sum()` returns the count of non-missing values in each column of the DataFrame `df`.
```

```python
df.notnull().sum().sum()#`df.notnull().sum().sum()` returns the total count of non-missing values across the entire DataFrame `df`.
```

```python
df.isnull()#`df.isnull()` returns a DataFrame of the same shape as `df`, with `True` for each `NaN` (missing) value and `False` for non-missing values.
```

```python
df.fillna(2) ##NAN er jaygay 2
```

```python
df.fillna({'Name':'None','Profit':0,'Growth':0})#`df.fillna({'Name':'None','Profit':0,'Growth':0})` fills the `NaN` (missing) values in the `Name`, `Profit`, and `Growth` columns of the DataFrame `df` with `'None'`, `0`, and `0`, respectively.
```

```python
df.fillna(method='ffill')#`df.fillna(method='ffill')` fills the `NaN` (missing) values in the DataFrame `df` by forward filling, meaning it replaces each `NaN` with the most recent non-missing value in the column.
```

```python
df.fillna(method='ffill',axis=0)#`df.fillna(method='ffill', axis=0)` fills the `NaN` (missing) values in the DataFrame `df` by forward filling along the rows (axis 0), meaning each `NaN` is replaced by the value from the previous row in the same column.
```

```python
df.fillna(method='ffill',axis=1)#`df.fillna(method='ffill', axis=1)` fills the `NaN` (missing) values in the DataFrame `df` by forward filling along the columns (axis 1), meaning each `NaN` is replaced by the value from the previous column in the same row.
```

```python
df.fillna(method='bfill',axis=0)#`df.fillna(method='bfill', axis=0)` fills the `NaN` (missing) values in the DataFrame `df` by backward filling along the rows (axis 0), meaning each `NaN` is replaced by the next non-missing value in the same column.
```

```python
df.fillna(0,limit=1)#`df.fillna(0, limit=1)` fills the first `NaN` (missing) value in each column of the DataFrame `df` with `0`, and stops after replacing one `NaN` per column, based on the `limit=1` parameter.
```
